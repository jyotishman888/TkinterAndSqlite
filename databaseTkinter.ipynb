```python
from tkinter import *
import sqlite3
```

```python
root = Tk()
root.title('My Database App')
root.geometry("300x400")
root.resizable(False,False)


#1 connecting to database
conn = sqlite3.connect('address_book.db')

#### creating a cursor 
c = conn.cursor()

#2 creating the table

# c.execute("""CREATE TABLE addresses (
# first_name text,
# last_name text,
# address text,
# city text,
# state text,
# zipcode integer
#  )
# """)


# conn.commit()

# #3 closing the connection 
# conn.close()

#7 Create the function




#9 Delete function

def delete():
    conn = sqlite3.connect('address_book.db')

    c = conn.cursor()
    
    
    c.execute("DELETE from addresses WHERE oid= " + delete_box.get())
    
    delete_box.delete(0, END)
    
    #commit changes
    conn.commit()
    #close connection
    conn.close()
    

def submit():
    conn = sqlite3.connect('address_book.db')

    c = conn.cursor() ## creating a cursor and always place it inside a function
    
    #8 Insert into table
    
    c.execute("INSERT INTO addresses VALUES(:f_name, :l_name, :address, :city, :state, :zipcode)",
             {
              'f_name': f_name.get(),
              'l_name': l_name.get(),
              'address': address.get(),
              'city': city.get(),
              'state': state.get(),
              'zipcode': zipcode.get()  
             })
    
    #commit changes
    conn.commit()
    #close connection
    conn.close()
 

    f_name.delete(0, END)
    l_name.delete(0, END)
    address.delete(0, END)
    city.delete(0, END)
    state.delete(0, END)
    zipcode.delete(0, END)
    

def query():
    conn = sqlite3.connect('address_book.db')

    c = conn.cursor()
    
    
    c.execute("SELECT *, oid FROM addresses")
    records = c.fetchall()
    # print(records)
    
    #Loops through results
    
    print_records = ''
    for record in records:
        print_records += str(record[0]) + " " + str(record[1]) + " | " + str(record[2]) +"\t\t" +str(record[6]) + "\n" 
        
    query_label = Label(root, text=print_records)
    query_label.grid(row=11, column=0, columnspan=2)
    
    
     #commit changes
    conn.commit()
    #close connection
    conn.close()
    



#4 created the text boxes
f_name = Entry(root, width=30)
f_name.grid(row=0, column=1, padx=20)

l_name = Entry(root, width=30)
l_name.grid(row=1, column=1, padx=20)

address = Entry(root, width=30)
address.grid(row=2, column=1, padx=20)

city = Entry(root, width=30)
city.grid(row=3, column=1, padx=20)

state = Entry(root, width=30)
state.grid(row=4, column=1, padx=20)

zipcode = Entry(root, width=30)
zipcode.grid(row=5, column=1, padx=20)

delete_box =Entry(root, width=30)
delete_box.grid(row=9, column=1)


#5 create text box label
f_name_label = Label(root, text="First Name")
f_name_label.grid(row=0, column=0)

l_name_label = Label(root, text="Last Name")
l_name_label.grid(row=1, column=0)

address_name_label = Label(root, text="Address")
address_name_label.grid(row=2, column=0)

city_name_label = Label(root, text="City")
city_name_label.grid(row=3, column=0)

state_name_label = Label(root, text="State")
state_name_label.grid(row=4, column=0)

zipcode_name_label = Label(root, text="Zipcode")
zipcode_name_label.grid(row=5, column=0)

delete_box_label = Label(root, text="ID Number")
delete_box_label.grid(row=9, column=0)


#6 create the button

submit_btn = Button(root, text="Add Record", command=submit) #after the button build the function
submit_btn.grid(row=6, column=0, columnspan=2, pady=10, padx=10, ipadx=100)

#9 Creating the query buttons to query the data from the database
query_btn = Button(root, text="Show Data", command=query)
query_btn.grid(row=7,column=0, columnspan=2, pady=10, padx=10, ipadx= 102)


delete_btn = Button(root, text="Delete Data", command=delete)
delete_btn.grid(row=10, column=0, columnspan=2, pady=10, padx=10, ipadx= 100)


#commit changes
conn.commit()
#close connection
conn.close()


root.mainloop()
```
